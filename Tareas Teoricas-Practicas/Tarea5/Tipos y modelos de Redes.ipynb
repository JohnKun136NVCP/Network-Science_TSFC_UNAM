# Modelo de Erd√∂s-Reny√≠(Redes aleatorias)
## Parte 1. 

1. El modelo de Erd√∂s Renyi es el t√≠pico modelo o el modelo cl√°sico que se utiliza a las llamadas  *redes aleatorias*. Esta consiste en una red se seleccionan un par de enodos y se a√±ade conexi√≥n entre ellos de manera *random* u aleatoria. Para ello se considera dos factores importantes:

    * El n√∫mero total de nodos $\displaystyle n$ el cual no se mueve (es fijo).
    * Existe una probabilidad $p$ en cual son los nodos arbitrarios que  est√°n conectados entre si.
    
2. Para la expresi√≥n an√°litica tenemos que para formar con $n$ nodos es $\displaystyle\frac{n(n-2)}{2}$ entonces la probabilidad de $p$ entre 2 nodos conectados es igual a $m$ que son las aristas de la red. Es decir:
\begin{equation}
    p = \frac{2m}{n(n-1)}
\end{equation}
Ahora para un nodo arbitrario $j$ tiene oportunidad de tener una conexi√≥n a lo m√°s $n-1$ de otros nodos. Entonces los $m$ enlaces que existen y el nodo $i$ est√° solamente a una $k$ de ellos. Es decir, la propabilidad $\displaystyle P(k)$ de que un nodo $i$ permanezca a estas $k$ es:
\begin{equation}
    P(k) = \left(n-1\atop k\right)p^{k}(1-p)^{(n-1)-k}
\end{equation}
En el caso donde la red sea muy grande se tiene que:
\begin{equation}
    P(k) = e^{-z}\frac{z^{k}}{k!} ; ~ ~  ~ \langle \textbf{k}\rangle = z = \frac{2m}{n} = p(n-1)
\end{equation}
En este caso $\langle\textbf{k}\rangle$ representa el grado promedio que la mayor√≠a de nodos tienen aproximandamente el mismo grado, con un valor cercano al grado promedio $\langle k\rangle$.

La distribuci√≥n de grado con 3 diferentes tipos de redes aleatorias.

Al aumentar la reconexi√≥n $p_{r}$ hay una transici√≥n de la red regular (*lattice*) a una red aleatoria como el modelo *Erd√∂s-R√©nyi*.

# Modelo de Barab√°si-Albert (Redes libres de escala).
## Punto 11.
1. El  modelo m√°s de Barab√°si-Albert es capaz de generar o crear un red de distribuci√≥n de grado que sigue una ley de pontencia, en el cual el n√∫mero de nodos aumeta conforme a $t$, es decir, es un modelo din√°mico en donde los nodos se a√±aden uno por uno (un modelo de crecimiento de red). Aqu√≠ la positividad que se establezca una conexi√≥n con un nodo ya existente es proporcional al grado de este √∫ltimo (enlace preferencial) $\displaystyle p_{i} = \frac{k_{i}}{\sum_{j}k_{j}}$.
2. Para $P(k)$ el modelo de Barab√°si-Albert es $P(k) = k^{-\gamma}$ donde genera redes cuya distribuci√≥n de grado se sigue de una ley d epotencia como la pone la expresi√≥n anterior. En el *Clustering Coefficient*  deca con el tam√±ao de la red, siguiendo aproximadamente una ley de potencias se tiene $C\sim n^{-\frac{3}{4}}$.
De manera an√°litica $P(k) = Ax^{-\gamma}\rightarrow \log{(P(x))} = \log{(Ax^{-\gamma})}\rightarrow *=\log(A)+\log(x^{-\gamma})=B+-\gamma\log{(x)} = -\gamma x$. Esto es igual a 
\begin{equation}
    y = -\gamma x +B
\end{equation}
Donde $y = \log(P(k))$ y $x = \log(k)$.

```python
#Punto 13.
def Distribution_barabasi(model):
    dictBarabasi = dict(nx.degree(model))
    sequencial = sorted(dictBarabasi.values(),reverse=True) #Ordenar solo los valores.
    fig = plt.figure(figsize=(8, 8))
    ax1=fig.add_subplot(221)
    histogramaBA = plt.hist(sequencial,bins=100)
    
    plt.title("Diagrama de distribuci√≥n logaritmica. Modelo: {}".format(model) )
    plt.xlabel("$K$")
    plt.ylabel("$P(K)$")
    
    ax2=fig.add_subplot(224)
    K = []
    for i in range(len(histogramaBA[1])-1):K.append(histogramaBA[1][i])
    plt.loglog(K,histogramaBA[0],'r.')
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("$\log(k)$")
    plt.ylabel("$\log(p(k))$")
    plt.title("Distribuci√≥n de grado en escala logaritmica. Modelo Barabasi.  ")
    return K,histogramaBA[0]
   # return histogramaBA
K,hist=Distribution_barabasi(_ba_1_1)
```

```python

#Con poly1d en numpy
x = np.log(K)
y = np.log(hist,out=np.zeros_like(hist),where=(hist!=0))
coeficiente = np.polyfit(x,y,1)
pfn = np.poly1d(coeficiente)
plt.xlabel("$log(k)$")
plt.ylabel("$log(P(k))$")
plt.plot(x,y,'r.')
plt.plot(x,pfn(x),'b--')
print(coeficiente)
```

```python
#Haciendo el ajuste a mano xd
def Pendiente(df):
  xy = [i*j for i,j in zip(df['x'],df['y'])] #xy
  xy =sum(xy) #suma de cada xy
  x = df['x'].sum()
  y = df['y'].sum()
  n = df.shape[0]
  x_sqr = sum([i**2 for i in df['x']])
  y_sqr = sum([i**2 for i in df['y']])
  sum_x_sqr = x**2
  m = (float(n)*xy-x*y)/(float(n)*x_sqr-sum_x_sqr)
  b = (y/n)-m*(x/n)
  return round(m,4),round(b,4)
def Graph(df,m,b):
  a = df.iloc[0]['x']
  z = df.iloc[df.shape[0]-1]['x']
  RangeL = np.linspace(a,z)
  fig, ax = plt.subplots(figsize=(8,8))
  ax.scatter(df['x'],df['y'],color='r',label="$\log(k)$ vs $\log(P(k))$")
  ax.plot(RangeL,m*RangeL+b,'-m', label = "$y=-1.1645x+4.496$")
  plt.grid(True)
  plt.xlabel('$\log(k)$')
  plt.tight_layout()
  plt.legend(loc='upper left')
  plt.ylabel('$\log(P(k))$')
  plt.show()
data = {'x':x,'y':y}
df = pd.DataFrame(data)
m,b=Pendiente(df)
Graph(df,m,b)
print(m,b)
```

# Partes opcionales.

<p>

</p>
1. Para $\langle L\rangle$ podemos verlo como:
\begin{equation}
    p_{L} = \left(~L ~max~\atop L\right)p^{L}(1-p)^{~L~max~-L}
\end{equation}
\begin{equation}
    \langle L\rangle = \sum_{L=0}^{L~max~} L\cdot p_{L} = (L~max~)\cdot p \longrightarrow \langle L\rangle = p\cdot\frac{N(N-1)}{2}
\end{equation}
Por lo que tenemos que $\langle K\rangle$
\begin{equation}
    \langle K\rangle =\frac{2\langle L\rangle}{N} = \frac{2pN(N-1)}{2N}\therefore \langle K\rangle = p(N-1)
\end{equation}
2. $p$ es la probabilidad que cualquier de los nodos en una red est√°n conectados, entonces el n√∫mero esperado de conexiones entre el $k_{i}$ vecinos de nodo $i$ es:
\begin{equation}
    \langle L\rangle = p\cdot\frac{N(N-1)}{2}
\end{equation}
El esperado local *clustering coeffcient* de nodo $i$ es:
\begin{equation}
    C =\frac{2\langle L\rangle}{N(N-1)} =\frac{2pN(N-1)}{N(N-1)2}  = p\therefore\langle C\rangle \simeq p
\end{equation}
