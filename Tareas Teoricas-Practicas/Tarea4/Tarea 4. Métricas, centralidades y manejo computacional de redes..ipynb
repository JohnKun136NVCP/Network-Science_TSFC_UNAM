1. Se define el $k_{i}$ es el n√∫mero de conexiones del nodo $i$ o tambi√©n como $k_{i}=g(i)=$ grado de $i$.
\begin{equation*}
    k_{i} = \sum_{j}A_{ij}
\end{equation*}
2. El coeficiente de agrupaci√≥n describe las conexiones m√°s cercanas a un nodo y sus vecinos. Se expresa de la siguiente manera:
\begin{equation*}
    C_{i} = \frac{~triangulos~conectados~al~nodo~ i}{~todos~los~posibles~triangulos~conectados~entre~si}
\end{equation*}
3. El *Between centrality* de nodo $i$ es el numero de caminos *geod√©sicos* entre cualquier dos nodos que pasan a trav√©s de $i$. Es decir, esto nos ayuda a determinar el nodo m√°s importante de una red ya que son los caminos que recorren o pasan en √©l.
4. *Shortest Path Length*: Es el camino m√°s corto entre dos nodos.
5. El diametro de la red se define como el maximo de las distancias $\langle l_{i,j}\rangle$.
6. Descrita como $\gamma_{i}$ que representa la lejan√≠a de un nodo del resto en promedio. Esto nos da informaci√≥n sobre otros nodos y su influencia uno con otro. Es la proximidad que mide qu√© tan lejos est√° uno nodo de todos los dem√°s.
\begin{equation*}
    \displaystyle\gamma = \frac{n-1}{\sum_{i}d_{ij}}
\end{equation*}
7. La excentrididad del nodo $e_{x}$ es el valor m√°ximo de las excentridades en toda la red.
8. Conocida como *Eigenvetor centrality* es la influencia m√°s acertada de una red. Es decir, es la suma de las puntaciones de sus vecinos es puntuaci√≥n relativa por cada vecino que tenga en la red.
9. La centralidad de *Katz* en el caso de las redes implicar√≠a que un solo nodo puede tener una centralidad de vector propio igual a 0 si el vecino de su √∫nica arista entrate tambi√©n tiene la centricidad de vector propio igual a 0, entonces tomando una cantidad random o arbitrar√≠a que se llama $\beta$ se tiene que
\begin{equation*}
    x_{i} = \alpha\sum_{j}A_{ij}x_{j}+\beta
\end{equation*}
En este caso  $\alpha$ y $\beta$ son constantes mayores que 0 que ser√°n el balance del valor de la centracidad $x_{i}$. El primer termino es la centracidad de vector propio y el segundo que es la $\beta$ a√±ade la centralidad haciendo el balance.
10. Apartir de la centralidad de Katz en la que el valor de un nodo botiene de sus vecinos sea proporcional a su dentralidad por el grado de salidad. En el caso los nodos apuntan a otros aprotan s√≥lo tienen una peque√±a centralidad. En este caso el nombre viene por Google porque es una base para clasificar sus p√°ginas de su bases de datos.

    \rho_{i} = \delta\sum_{j}\frac{\rho_{j}}{k^{out}_{j}} + \frac{1-\delta}{n}
\end{equation*}
